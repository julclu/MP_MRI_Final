```python
import os 
import glob
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
path = pd.read_csv('po1_preop_recur_path_jan2019.csv')
path.shape
```

```python
## remove the tumor_cell_evaluations that don't have entries
path_withoutNaN = path.dropna(subset = ['tumor_cell_evaluation'])
path_withoutNaN.shape
```

```python
## remove duplicates for t_number
path_withoutNaN_noDup = path_withoutNaN.drop_duplicates('t_number')
```

```python
path_withoutNaN_noDup.shape
```

```python
## remove Grade II that are in there accidentally: 
path_withoutNaN_noDup_noGII = path_withoutNaN_noDup[path_withoutNaN_noDup.current_hist_grade != "Grade II"]
```

```python
path_withoutNaN_noDup_noGII.shape
```

```python
path_withoutNaN_noDup_noGII.t_number = path_withoutNaN_noDup_noGII.t_number.astype(int)
```

```python
path_withoutNaN_noDup_noGII.shape
```

```python
path_withoutNaN_noDup_noGII.head()
```

```python
path_withoutNaN_noDup_noGII.b_number = [bnum  if len(str(bnum)) == 4 else "0"+str(bnum)[0:len(bnum)] for bnum in path_withoutNaN_noDup_noGII.b_number.astype(str)]
```

```python
path_withoutNaN_noDup_noGII.b_number = "b"+path_withoutNaN_noDup_noGII.b_number
```

```python
path_withoutNaN_noDup_noGII.head()
```

```python
path_withoutNaN_noDup_noGII.t_number = "t"+path_withoutNaN_noDup_noGII.t_number.astype(str)
```

```python
path_withoutNaN_noDup_noGII.head()
```

```python
## only want to keep the bnums that are in the /archived folder, else won't be able to process them correctly: 
bnums_in_dir = os.listdir('/data/RECglioma/archived/')
```

```python
bnum_tnum_list = path_withoutNaN_noDup_noGII[['b_number', 't_number']]
bnum_tnum_list.head()
```

```python
bnum_tnum_list_indir = bnum_tnum_list[bnum_tnum_list['b_number'].isin(bnums_in_dir) ]
```

```python
bnum_tnum_list_indir.shape
```

```python
bnum_tnum_list_indir['DUMMY']="DUMMY"
```

```python
bnum_tnum_list_indir.head()
```

```python
bnum_tnum_list_indir.to_csv('po1_preop_recur_process_list.125.csv', index = False, header =False )
```
