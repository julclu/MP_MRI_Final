```python
diffu2000.columns
```

```python
diffu2000.columns = ['id_value', 'biopsy', 'adc.2', 'fa.2', 'nadc.2', 'nfa.2', 'nev1.2',
       'nev2.2', 'nev3.2', 'nevrad.2']
```

```python
perf = pd.read_csv('po1_preop_recur_perf_biopsy.csv')
```

```python
perf.columns
```

```python
perf.columns = ['bnum', 'id_value', 'biopsy', 'phn_nlin', 'cbvn_nlin', 'recov_nlin',
       'phn_npar', 'recov_npar', 'recovn_npar']
```

```python
spec = pd.read_csv('oldP01_spec_stats.biopsy.csv', index_col = 0)
spec.head()
spec.columns
```

```python
spec.columns = ['id_value', 'biopsy', 'vol/fvox', 'cho', 'cre', 'naa', 'lip', 'lac',
       'xschn', 'xsccr', 'xscrn', 'cni', 'ccri', 'crni', 'ncho', 'ncre',
       'nnaa', 'nlip', 'nlac', 'nxschn', 'nxsccr', 'nxscrn']
```

```python
ad = pd.merge(anat, diffu, 'outer', on = ['id_value', 'biopsy'])
ad.shape
```

```python
add = pd.merge(ad, diffu2000, 'outer', on = ['id_value', 'biopsy'])
```

```python
adp = pd.merge(add, perf, 'outer', on = ['id_value', 'biopsy'])
adp.shape
```

```python
adps = pd.merge(adp, spec, 'outer', on = ['id_value', 'biopsy'])
adps.shape
```

```python
adps = adps.groupby("biopsy", as_index=False).first()
```

```python
len(adps.columns)
```

```python
adps.columns
```

```python
cols_of_interest = ['biopsy', 
                   'id_value', 
                   'nt1c', 
                   'nt1v', 
                   'nfse', 
                   'nfl', 
                   'nt1d',
                   'nadc.1', 'nfa.1', 
                   'nev1.1', 'nev2.1', 'nev3.1', 
                   'nevrad.1',
                   'nadc.2', 'nfa.2', 
                   'nev1.2', 'nev2.2', 'nev3.2', 
                   'nevrad.2',
                    'phn_nlin', 'cbvn_nlin', 'recov_nlin', 
                   'phn_npar','recov_npar', 'cni', 'ccri', 'crni', 
                   'ncho', 'ncre', 'nnaa', 'nlip', 'nlac']
```

```python
old_data_final = adps[cols_of_interest]
```
