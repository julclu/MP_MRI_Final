```python
# sci python and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn.preprocessing import StandardScaler

# model selection and performance
from sklearn.model_selection import train_test_split,cross_val_score,KFold
from sklearn.metrics import accuracy_score, r2_score, roc_auc_score, classification_report,confusion_matrix
from sklearn.feature_selection import SelectKBest,chi2, f_classif

#models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

# decorations
from IPython.display import YouTubeVideo

import warnings
warnings.filterwarnings('ignore')

np.random.seed(2)
import random
import seaborn as sns 
```

```python
data = pd.read_csv("../../AnnotateData/recgli_annotation10_CELorNELorUNK.csv")
print("Your data has",data.shape[0], "rows, and ", data.shape[1], "columns")
data.head()
```

```python
labels = data.outcome
data.shape
```

```python
data.columns
```

```python
features = ['vialid', 't_number_y', 'nadc.1', 'nfa.1',
       'cbvn_nlin', 'recov_npar',
       'ccri', 'cni', 'ncho', 'ncre', 'nnaa', 'newdata', 'CELorNEL', 'outcome']
```

```python
data = data[features]
```

```python
data.head()
```

```python
data_cel = data.loc[data['CELorNEL'] == "CEL"]
```

```python
data_nel = data.loc[data['CELorNEL'] == "NEL"]
```

```python
data = data_cel.append(data_nel)
```

```python
data.CELorNEL.value_counts()
```

```python
data.shape
```

```python
data_txe = data.loc[data['outcome']=="txe"]
data_rHGG = data.loc[data['outcome']=="rHGG"]
```

```python
data = data_txe.append(data_rHGG)
```

```python
data.shape
```

## Encode for missingness: 

```python
data['hasDiffu'] = [0 if x else 1 for x in data['nadc.1'].isnull()]
data['hasPerf'] = [0 if x else 1 for x in data['cbvn_nlin'].isnull() ]
data['hasSpec'] = [0 if x else 1 for x in data['cni'].isnull() ]
```

## adding in the CEL, outcome yes/no binary 

```python
data['in_CEL'] = pd.get_dummies(data['CELorNEL'])['CEL']
```

```python
del data['CELorNEL']
```

```python
data['outcome_bin']=[0 if x == "txe" else 1 for x in data['outcome']]
```

```python
data.head()
```

## splitting into train/test (by patient), ensuring good stratification

```python
## what is current stratification? 
data.outcome.value_counts()['txe']/(data.outcome.value_counts()['txe'] + data.outcome.value_counts()['rHGG'])
```

```python
unique_patients = data['t_number_y'].unique().tolist()
```

```python
random.seed(1)
random.shuffle(unique_patients)
```

```python
train_tnums = unique_patients[0:round(len(unique_patients)*.8)]
test_tnums = [x for x in unique_patients if x not in train_tnums]
```

```python
data_train = data[data['t_number_y'].isin(train_tnums)]
data_train.outcome.value_counts()['txe']/(data_train.outcome.value_counts()['txe'] + data_train.outcome.value_counts()['rHGG'])

```

```python
data_test = data[data['t_number_y'].isin(test_tnums)]
data_test.outcome.value_counts()['txe']/(data_test.outcome.value_counts()['txe'] + data_test.outcome.value_counts()['rHGG'])
```

```python
y_train = data_train.outcome_bin
del data_train["outcome_bin"]

y_test = data_test.outcome_bin
del data_test['outcome_bin']
```

```python
del data_train["outcome"]
del data_test['outcome']
```

## normalizing data

```python
for col in ['nadc.1', 'nfa.1',
       'cbvn_nlin', 'recov_npar',
       'ccri', 'cni', 'ncho', 'ncre', 'nnaa']: 
    mean_col = data_train[col].mean()
    sd_col = data_train[col].std()
    data_train[col] = (data_train[col]-mean_col)/sd_col
    data_test[col] = (data_test[col]-mean_col)/sd_col
```

## use median imputation for both training and testing: 

```python
data_test.loc[data_test['nadc.1'].isnull(), ['nadc.1']] = data_train['nadc.1'].median()
data_train.loc[data_train['nadc.1'].isnull(), ['nadc.1']] = data_train['nadc.1'].median()
```

```python
data_test.loc[data_test['nfa.1'].isnull(), ['nfa.1']] = data_train['nfa.1'].median()
data_train.loc[data_train['nfa.1'].isnull(), ['nfa.1']] = data_train['nfa.1'].median()
```

```python
data_test.loc[data_test['cbvn_nlin'].isnull(), ['cbvn_nlin']] = data_train['cbvn_nlin'].median()
data_train.loc[data_train['cbvn_nlin'].isnull(), ['cbvn_nlin']] = data_train['cbvn_nlin'].median()
```

```python
data_test.loc[data_test['recov_npar'].isnull(), ['recov_npar']] = data_train['recov_npar'].median()
data_train.loc[data_train['recov_npar'].isnull(), ['recov_npar']] = data_train['recov_npar'].median()
```

```python
data_test.loc[data_test['cni'].isnull(), ['cni']] = data_train['cni'].median()
data_train.loc[data_train['cni'].isnull(), ['cni']] = data_train['cni'].median()
```

```python
data_test.loc[data_test['ccri'].isnull(), ['ccri']] = data_train['ccri'].median()
data_train.loc[data_train['ccri'].isnull(), ['ccri']] = data_train['ccri'].median()
```

```python
data_test.loc[data_test['ncho'].isnull(), ['ncho']] = data_train['ncho'].median()
data_train.loc[data_train['ncho'].isnull(), ['ncho']] = data_train['ncho'].median()
```

```python
data_test.loc[data_test['ncre'].isnull(), ['ncre']] = data_train['ncre'].median()
data_train.loc[data_train['ncre'].isnull(), ['ncre']] = data_train['ncre'].median()
```

```python
data_test.loc[data_test['nnaa'].isnull(), ['nnaa']] = data_train['nnaa'].median()
data_train.loc[data_train['nnaa'].isnull(), ['nnaa']] = data_train['nnaa'].median()
```

## assigning x_train, x_test, y_train, y_test

```python
x_train, x_test, y_train, y_test = data_train, data_test, y_train, y_test
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

## modeling

```python
models = []
models.append(('KNN', KNeighborsClassifier()))
models.append(('SVC', SVC()))
models.append(('LR', LogisticRegression()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('GNB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('GB', GradientBoostingClassifier()))
```

```python
names = []
scores = []
for name, model in models:
    
    kfold = KFold(n_splits=5, random_state=10) 
    score = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy').mean()
    
    names.append(name)
    scores.append(score)
kf_cross_val = pd.DataFrame({'Name': names, 'Score': scores})
print(kf_cross_val)
```

```python
names = []
scores = []
for name, model in models:
    
    kfold = KFold(n_splits=5, random_state=10) 
    score = cross_val_score(model, x_train_tst, y_train_tst, cv=kfold, scoring='accuracy').mean()
    
    names.append(name)
    scores.append(score)
kf_cross_val = pd.DataFrame({'Name': names, 'Score': scores})
print(kf_cross_val)
```

```python
names = []
scores = []
for name, model in models:
    
    kfold = KFold(n_splits=5, random_state=10) 
    score = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
    
    names.append(name)
    scores.append(score)
```

```python
names_tst = []
scores_tst = []
for name, model in models:
    
    kfold = KFold(n_splits=5, random_state=10) 
    score = cross_val_score(model, x_train_tst, y_train_tst, cv=kfold, scoring='accuracy')
    
    names_tst.append(name)
    scores_tst.append(score)
```
